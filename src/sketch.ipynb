## Signal Shift

```julia
s = impseq(-10, 10);
sshifted = sigshift(s, 5)
```

```julia
p1 = plot(s.n, s.A, line =:stem, marker=:o)
p2 = plot(sshifted.n, sshifted.A, line =:stem, marker=:o)
plot(p1,p2, layout=(1,2))
```

```julia
s = stepseq(-10, 10);
sshifted = sigshift(s, 5)
```

```julia
p1 = plot(s.n, s.A, line =:stem, marker=:o)
p2 = plot(sshifted.n, sshifted.A, line =:stem, marker=:o)
plot(p1,p2, layout=(1,2))
```

```julia
s = stepseq(-10, 10);
sshifted = sigshift(s, -5)
```

## Signal add

```julia
s1 = impseq(-10, 10);
s2 = stepseq(-5, 15);

```

```julia
sigadd(s1, s2)
```

```julia
p1 = plot(s1.n, s1.A, line =:stem, marker=:o)
p2= plot(s2.n, s2.A, line =:stem, marker=:o)
p3 = plot(sigadd(s1, s2).n, sigadd(s1, s2).A, line =:stem, marker=:o)
plot(p1,p2,p3, layout=(3,1), xlims=(-10, 15))
```

```julia
s1 = stepseq(-15, 2);
s1 = sigshift(s1, -10)
```

```julia
s2  = stepseq(-1, 40);
s2 = sigshift(s2, 10)
```

```julia
p1 = plot(s1.n, s1.A, line =:stem, marker=:o, xlims=(-25,50))
p2= plot(s2.n, s2.A, line =:stem, marker=:o, xlims=(-25,50))
p3 = plot(sigadd(s1, s2).n, sigadd(s1, s2).A, line =:stem, marker=:o, xlims=(-25,50))
plot(p1,p2,p3, layout=(3,1))
```

# Signal Multiply

### Multiplication can work as a "mask" , in order to let a "window" of the signal pass.

```julia
s1 = stepseq(-10, 10);
s1 = sigshift(s1, -5)
```

```julia
plot(s1.n,s1.A, line =:stem, marker=:o)
```

```julia
s2 = signal(0.0, -15, 15)
```

```julia
s3 = sigadd(s1,s2)
```

```julia
p1 = plot(s1.n, s1.A, line =:stem, marker=:o, xlims=(-15,15))
p2 = plot(s2.n, s2.A, line =:stem, marker=:o, xlims=(-15,15))
p3 = plot(s3.n, s3.A, line =:stem, marker=:o, xlims=(-15,15))
plot(p1,p2,p3, layout=(3,1))
```

```julia
s4 = signal(1.0, -15, 15)
```

```julia
s5 = signal(0.0, -15, 15)
```

```julia
s5.A = rand.().*s4.A
```

```julia
p5 = plot(s5.n, s5.A, line =:stem, marker=:o, xlims=(-15,15), ylims=(0,1))
```

```julia
s6 = sigmult(s5, s3)
```

```julia
p1 = plot(s3.n, s3.A, line =:stem, marker=:o, xlims=(-15,15))
p2 = plot(s5.n, s5.A, line =:stem, marker=:o, xlims=(-15,15))
p3 = plot(s6.n, s6.A, line =:stem, marker=:o, xlims=(-15,15))
plot(p1,p2,p3, layout=(3,1))
```

# Signal Scaling

```julia
s1 = sigrand(-20,20)
s2 = sigscale(s1, 0.3) # scale by 0.3
```

```julia
p1 = plot(s1.n, s1.A, line=:stem, color=:red, marker=:o, xlims=(-20,20), ylims=(0.0,1.0))
p2 = plot(s2.n, s2.A, line=:stem, color=:blue, marker=:o, xlims=(-20,20), ylims=(0.0,1.0))
plot(p1,p2, layout=(2,1))
```
