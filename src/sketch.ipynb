```julia
using MySignalProcessing
using Plots
```

## UNIT SAMPLE SEQUENCE

```julia
s = impseq(-10, 10);
```

```julia
plot(s.n, s.A, line =:stem, marker=:o)
```

## UNIT STEP SEQUENCE

```julia
s = stepseq(-10, 10);
```

```julia
plot(s.n, s.A, line =:stem, marker=:o)
```

## Signal Shift

```julia
s = impseq(-10, 10);
sshifted = sigshift(s, 5)
```

```julia
p1 = plot(s.n, s.A, line =:stem, marker=:o)
p2 = plot(sshifted.n, sshifted.A, line =:stem, marker=:o)
plot(p1,p2, layout=(1,2))
```

```julia
s = stepseq(-10, 10);
sshifted = sigshift(s, 5)
```

```julia
p1 = plot(s.n, s.A, line =:stem, marker=:o)
p2 = plot(sshifted.n, sshifted.A, line =:stem, marker=:o)
plot(p1,p2, layout=(1,2))
```

```julia
s = stepseq(-10, 10);
sshifted = sigshift(s, -5)
```

```julia
p1 = plot(s.n, s.A, line =:stem, marker=:o)
p2 = plot(sshifted.n, sshifted.A, line =:stem, marker=:o)
plot(p1,p2, layout=(1,2))
```

## Signal add

```julia
s1 = impseq(-10, 10);
s2 = stepseq(-5, 15);

```

```julia
sigadd(s1, s2)
```

```julia
p1 = plot(s1.n, s1.A, line =:stem, marker=:o)
p2= plot(s2.n, s2.A, line =:stem, marker=:o)
p3 = plot(sigadd(s1, s2).n, sigadd(s1, s2).A, line =:stem, marker=:o)
plot(p1,p2,p3, layout=(3,1), xlims=(-10, 15))
```

```julia
s1 = stepseq(-15, 2);
s1 = sigshift(s1, -10)
```

```julia
s2  = stepseq(-1, 40);
s2 = sigshift(s2, 10)
```

```julia
p1 = plot(s1.n, s1.A, line =:stem, marker=:o, xlims=(-25,50))
p2= plot(s2.n, s2.A, line =:stem, marker=:o, xlims=(-25,50))
p3 = plot(sigadd(s1, s2).n, sigadd(s1, s2).A, line =:stem, marker=:o, xlims=(-25,50))
plot(p1,p2,p3, layout=(3,1))
```
