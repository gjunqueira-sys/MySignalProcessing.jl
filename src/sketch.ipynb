## Signal Shift

## Most of the Functions are designed for be composable

```julia
s = impseq(-10, 10);
sshifted = s |> sigshift(5)
```

```julia
p1 = plot(s.n, s.A, line =:stem, marker=:o)
p2 = plot(sshifted.n, sshifted.A, line =:stem, marker=:o)
fig3 = plot(p1,p2, layout=(1,2))
savefig(fig3, "impulse_shifted.png")
display(fig3)
```

# Signal Multiply

### Multiplication can work as a "mask" , in order to let a "window" of the signal pass.

```julia

s1 = stepseq(-10, 10) |> sigshift(-5)
```

```julia
plot(s1.n,s1.A, line =:stem, marker=:o)
```

```julia
s2 = signal(0.0, -15, 15)
```

```julia
s3 = sigadd(s1,s2)
```

```julia
p1 = plot(s1.n, s1.A, line =:stem, marker=:o, xlims=(-15,15))
p2 = plot(s2.n, s2.A, line =:stem, marker=:o, xlims=(-15,15))
p3 = plot(s3.n, s3.A, line =:stem, marker=:o, xlims=(-15,15))
plot(p1,p2,p3, layout=(3,1))
```

```julia
s4 = signal(1.0, -15, 15)
```

```julia
s5 = signal(0.0, -15, 15)
```

```julia
s5.A = rand.().*s4.A
```

```julia
p5 = plot(s5.n, s5.A, line =:stem, marker=:o, xlims=(-15,15), ylims=(0,1))
```

```julia
s6 = sigmult(s5, s3)
```

```julia
p1 = plot(s3.n, s3.A, line =:stem, marker=:o, xlims=(-15,15))
p2 = plot(s5.n, s5.A, line =:stem, marker=:o, xlims=(-15,15))
p3 = plot(s6.n, s6.A, line =:stem, marker=:o, xlims=(-15,15))
plot(p1,p2,p3, layout=(3,1))
```

# Signal Scaling

```julia
s1 = sigrand(-20,20)
s2 = s1 |> sigscale(0.3) # scale by 0.3
```

```julia
p1 = plot(s1.n, s1.A, line=:stem, color=:red, marker=:o, xlims=(-20,20), ylims=(0.0,1.0))
p2 = plot(s2.n, s2.A, line=:stem, color=:blue, marker=:o, xlims=(-20,20), ylims=(0.0,1.0))
plot(p1,p2, layout=(2,1))
```

# Signal Folding

```julia
s = realexp(1.1, -50, 50);
```

```julia
plot(s.n, s.A, line =:stem, marker=:o)
```

```julia
s2 = s |> sigfold
```

```julia

p1 = plot(s.n, s.A, line =:stem, marker=:o)
p2 = plot(s2.n, s2.A, line =:stem, marker=:o)
plot(p1,p2, layout=(2,1))
```

```julia
plot(s.n, s.A, line =:stem, marker=:o)
plot!(s2.n, s2.A, line =:stem, marker=:o)
```

```julia
s = sigrand(5,7)
s1 =  s |> sigfold
```

```julia
plot(s.n, s.A, line =:stem, marker=:o)
plot!(s1.n, s1.A, line =:stem, marker=:o)
```
