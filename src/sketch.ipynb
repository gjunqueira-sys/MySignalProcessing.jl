```julia
mutable struct signal2 

    A::Vector{Real} #Amplitude of the signal
    n::Vector{Real} #horizontal axis (discrete) of signal (in samples)

    function signal2(A::Real, n::Int) # construction of signal with amplitude A and n samples starting from 0.
        A = A .*ones(n)
        n = [x for x in range(0, length=n)]
        new(A, n)
    end

    function signal2(A::Real, start::Int, stop::Int)
        
        n = [x for x in range(start, stop)]
        A = A .*ones(length(n))
        new(A, n)
    end


    function signal2(A::Vector, n::Vector)
        @assert length(A) == length(n) "lengths of A and n must be equal"
        new(A, n)
    end

    



end
```

```julia
function siggen((A::Vector))
    return A -> signal2(A,n::Vector)
end
```
